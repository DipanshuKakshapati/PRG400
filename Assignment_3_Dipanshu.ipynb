#### 1. Write a Python program that demonstrates the use of an `asyncio` event loop to run asynchronous tasks.

```python
import asyncio
```

```python
async def main():
    print("Starting the program...")
    await asyncio.sleep(1)  # pause for 1 second
    print("Program has ended.")
```

```python
# directly await the coroutine
await main()
```

#### 2. Write a Python program that demonstrates the use of a callback function.

```python
def process_data(data, callback):
    print(f"Processing data: {data}")
    result = data.upper()  # simulate data processing by converting to uppercase
    callback(result)  # trigger the callback with the result
```

```python
def display_result(processed_data):
    print(f"Processed Data: {processed_data}")
```

```python
# main program
data = "hello"
process_data(data, display_result)
```

#### 3. Write a Python program using `asyncio` that demonstrates concurrent execution of multiple tasks with different wait times.

```python
async def task1():
    print("Task 1 started")
    await asyncio.sleep(3)  # wait for 3 seconds
    print("Task 1 completed")
```

```python
async def task2():
    print("Task 2 started")
    await asyncio.sleep(2)  # wait for 2 seconds
    print("Task 2 completed")
```

```python
async def task3():
    print("Task 3 started")
    await asyncio.sleep(1)  # wait for 1 second
    print("Task 3 completed")
```

```python
async def main():
    # run all tasks concurrently
    await asyncio.gather(task1(), task2(), task3())
```

```python
# directly await the coroutine
await main()
```
