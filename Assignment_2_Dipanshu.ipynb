##### 15. Write a generator function `countup(start, end)` that yields numbers from `start` to `end` (inclusive). 

```python
def countup(start, end):
    for num in range(start, end + 1):
        yield num

for number in countup(2, 10):
    print(number)
```

##### 16. Write a generator function `even_numbers(n)` that yields the first `n` even numbers.

```python
def even_numbers(n):
    num = 2
    for _ in range(n):
        yield num
        num += 2

for number in even_numbers(10):
    print(number)
```

##### 18. Write a decorator named `repeat` that takes an integer `n` as an argument.The decorator should repeat the execution of the decorated function `n` times. Use the `repeat` decorator to modify a function called `print_message` that prints a custom message. Decorate the `print_message` function with your `repeat` decorator, passing `5` as the argument.Verify that the function is called 5 times and that the message is printed each time.

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(5)
def print_message(message="Hello!"):
    print(message)

print_message()
```
